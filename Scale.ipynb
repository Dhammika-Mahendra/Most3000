```python
import matplotlib.pyplot as plt
import string


file_path = 'Oxford 3000.txt' 
with open(file_path, 'r') as file:
    words = file.read().split()


all_letters = ''.join(words)


a_positions = [i for i, letter in enumerate(all_letters) if letter == 'a']


plt.figure(figsize=(12, 4))
plt.scatter(a_positions, [1] * len(a_positions), marker='|', color='blue', s=1000)


plt.yticks([])
plt.xlabel('Position in Concatenated String')
plt.title("Positions of 'A' in Concatenated String of All Words")
plt.show()
```

```python
import matplotlib.pyplot as plt
import string

file_path = 'Oxford 3000.txt' 
with open(file_path, 'r') as file:
    words = file.read().split()

all_letters = ''.join(words)

letter_positions = {letter: [i for i, l in enumerate(all_letters) if l == letter] for letter in string.ascii_lowercase}

fig, axs = plt.subplots(26, 1, figsize=(20, 40), sharex=True)

for ax, letter in zip(axs, string.ascii_lowercase):
    positions = letter_positions[letter]
    ax.scatter(positions, [1] * len(positions), marker='|', color='blue', s=100)
    ax.set_yticks([])
    ax.set_title(f"'{letter.upper()}'")
    ax.set_xlabel('Position')

plt.tight_layout()
plt.show()
```
